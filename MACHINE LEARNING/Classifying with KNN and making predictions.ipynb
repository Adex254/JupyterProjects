```python
# Classifying Data with Kneighbors Classifier and making predictions

# The required steps are:
# Step 1: impport libraries
# Step 2: load the dataset in
# Step 3: prepare features and labels
# Step 4: split train/test
# Step 5: Build your KNN model
# Step 6: Train, more accurately store in KNN
# Step 7: Predict
# Step 8: Evaluate
```

```python
# Ex 1 with iris dataset

# Step 1: import the libraries

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Step 2: load the dataset

iris = load_iris()

# Step 3: prepare the features and labels
X = iris.data
Y = iris.target

# Step 4: split the data into train and test

X_train, X_test, Y_train, Y_test = train_test_split(
    X, 
    Y,
    test_size = 0.3,
    stratify = Y,
    random_state = 1
)

# Step 5: Build your KNN model

knn = KNeighborsClassifier(3)   # picks the closest 5 members by default
# knn = KNeighborsClassifiers(n_neighbors = 3)  for three nearest neighbours

# Step 6: Train/store  the data in knn

knn.fit(X_train, Y_train) # giving it questions and answers

# Step 7: Predict

Y_pred = knn.predict(X_test)  # predicting value of X

# Step 8: Evaluate accuracy

accuracy = accuracy_score(Y_test, Y_pred) # compare the prediction with the actual data

print("Accuracy: ", accuracy)

 


```

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

# load in the iris dataset

iris = load_iris()
# print(iris)

# Split data into features and labels

features = iris.data
labels = iris.target

#print(features)
#print(labels)

# Split the data into train & test

features_train, features_test, labels_train, labels_test = train_test_split(
    features,
    labels,
    test_size = 0.2,
    random_state = 1
)

# Use the KNN Classifier to "train" the data

knn = KNeighborsClassifier(n_neighbors = 3)  # use 3 of the nearest neighbors

knn.fit(features_train, labels_train)  # train the model with features_train and labels train

# make predictions

predictions = knn.predict(features_test)
#print(predictions)

# estimate the accuracy

accuracy = accuracy_score(labels_test, predictions)
print("Accuracy: ", accuracy)



```

```python
# Accepting input from users


from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# load the dataset in

iris = load_iris()

# Split the data into features and labels

features = iris.data
labels = iris.target

# Split the data into train and test

features_train, features_test, labels_train, labels_test = train_test_split(
    features,
    labels,
    test_size = 0.25,
    random_state = 1,
    stratify = labels
)

# Build the knn classifier
knn = KNeighborsClassifier(n_neighbors = 3)

# Train the knn with the data

knn.fit(features_train, labels_train)

# new flower to make predictions
sepal_length = int(input("Enter the sepal length: "))
sepal_width = int(input("Enter the sepal sidth: "))
petal_length = int(input("Enter the petal length: "))
petal_width = int(input("Enter the petal width: "))

new_flower = [[sepal_length, sepal_width, petal_length, petal_width]]

# make predictions
prediction = knn.predict(new_flower)

# display the predictions 
print("Prediction class index: ", prediction[0])
print("Prediction name is: ", iris.target_names[prediction][0])

# check how accurate our model is
test_predictions = knn.predict(features_test)
accuracy = accuracy_score(labels_test, test_predictions)
print("Accuracy is: ", accuracy)
```
