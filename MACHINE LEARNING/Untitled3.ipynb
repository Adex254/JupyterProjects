```python
# import the important libraries

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


# Load the dataset

diabetes_df = pd.read_csv(r"C:\Users\adead\Downloads\csv dataset for practice\diabetes.csv")


# Split the dataset into features and labels

features = diabetes_df.drop(columns = ['Outcome'])
labels = diabetes_df['Outcome']


# Split the data into training and testing

features_train, features_test, labels_train, labels_test = train_test_split(
    features,
    labels,
    test_size = 0.2,
    random_state = 42,
    stratify = labels
)


# Build and 'train' the KNN model

knn = KNeighborsClassifier(n_neighbors = 2) # use the 5 nearest neighbors
knn.fit(features_train, labels_train)  # training


# make predictions
Pregnancies = int(input("Enter pregnancies: "))
Glucose = int(input("Enter glucose: "))
BloodPressure = int(input("Enter Blood pressure: "))
SkinThickness = int(input("Enter skin thickness: "))
Insulin = int(input("Enter insulin: "))
BMI = float(input("Enter BMI: "))
DiabetesPedigreeFunction = float(input("Enter diabetes pedigree function: "))
Age = int(input("Enter age: "))


test_data = [Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age]
prediction = knn.predict([test_data])

test_predictions = knn.predict(features_test)
# check accuracy

accuracy = accuracy_score(labels_test, test_predictions)

print("Prediction class: ", prediction[0])
if prediction[0] == 1:
    print("Outcome: Diabetes")
else:
    print("Outcome: No diabetes")
    
 
print("Accuracy on test data: ", accuracy)
```

```python
# using knn on sports dataset

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# load the dataset in

sports_df = pd.read_csv(r"C:\Users\adead\Downloads\csv dataset for practice\sports.csv.zip")
#print(sports_df)
#print(sports_df.shape)

# Split into features and labels

features = sports_df.drop(columns = ['sports'])
labels = sports_df['sports']

print(features.head())

# Split into training and features

features_train, features_test, labels_train, labels_test = train_test_split(
    features,
    labels,
    test_size = 0.15,
    random_state = 42,
    stratify = labels
)

# Build and train the KNN model 

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(features_train, labels_train)

#  make predictions

prediction = knn.predict(features_test)
print("Prediction class: ", prediction[0])

# check accuracy

accuracy = accuracy_score(labels_test, prediction)

print("Accuracy is: ", accuracy)


```
