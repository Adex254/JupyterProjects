```python
import pandas as pd
import numpy as np

excel_file = pd.read_excel(r"C:\Users\adead\Downloads\excel dataset for practice\Netflix-Movies-Sample-Data.xlsx", header=5)
df1 = pd.DataFrame(excel_file)

df1.drop("Unnamed: 0", axis=1, inplace=True) 
df1
```

```python
# picking rows using location
# SHIFT + TAB
# [ ] RETURNS A SERIES
# [ ] RETURNS A DF

df1.loc[41]
```

```python
# USING iloc[]
# for the rows
# iloc[row index]
df1.iloc[33]

# multiple rows
df1.iloc[[31, 33]]
```

```python
# picking a specific point
# use loc[]
# loc[rows, columns]

df1.loc[43, 'Name']
```

```python
df1.loc[27, 'Category']
```

```python
# several points
# df.loc[[row1, row2, row3, ..., row n], [col1, col2, col3, ..., col n]]

df1.loc[[22, 3, 37], ['Name', 'Duration', 'Category']]
```
