```python
# 3 most common methods of adding new columns
# 1. From lists
# 2. using the insert()
# 3. using the assign()

```

```python
import pandas as pd
import numpy as np

excel_file = pd.read_excel(r"C:\Users\adead\Downloads\excel dataset for practice\Financial Sample.xlsx")
df = pd.DataFrame(excel_file)
df1 = df.head() 
df1
```

```python
new_col = [20, 20, 15, 15, 15]
df1['Revenue per Unit'] = new_col
df1
```

```python
# Adding default values
df1['Good business'] = [True] * len(df1)
df1
```

```python
# Add NaN value
df1['Transaction ID'] = [np.nan] * len(df1)
df1
```

```python
# 2. insert()
# for any position
df1.insert(loc = 1, column = "Channel", value= ["Agent A"] * len(df1), allow_duplicates=False)
df1 
```

```python
# 3. assign()
df2 = df1.assign(Region = ["Africa"] * len(df1))
df2
```
