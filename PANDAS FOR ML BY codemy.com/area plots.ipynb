```python
import pandas as pd
import numpy as np



csv_file = pd.read_csv(r"C:\Users\adead\Downloads\csv dataset for practice\diabetes.csv")
df = pd.DataFrame(csv_file)
df.info()
df.describe()
```

```python
# Area plot
df.plot(kind="area")
```

```python
# Shading - alpha
df.plot(kind = "area", alpha = 0.75)
```

```python
# legend
df.plot(kind = "area", alpha = 0.75, legend = False)
```

```python
# title 
df.plot(kind = "area", alpha = 0.75, legend = True, title = "My Area Plot")
```

```python
# x and y-labels
df.plot(kind = "area", alpha = 0.75, legend = True, title = "My Area Plot", xlabel = "", ylabel = "Frequency")
```

```python
# table
df.plot(kind="area", legend = True, alpha = 0.75, title = "My Area Plot", xlabel = "", ylabel = "Frequency", table = True)
```

```python
# figsize 
df.plot(kind = "area", alpha = 0.75, legend = True, title = "My Area Plot", xlabel = "", ylabel = "Frequency", figsize = (12, 6))
```

```python
# area plot for one column
df['Pregnancies'].plot(kind = "area", alpha = 0.75, legend = True, title = "My Area Plot", xlabel = "", ylabel = "Frequency", figsize = (12, 6))
```

```python
# multiple cols
df[['BMI', 'Age', 'Outcome']].plot(kind = "area", alpha = 0.75, legend = True, title = "My Area Plot", xlabel = "", ylabel = "Frequency", figsize = (12, 6))
```
