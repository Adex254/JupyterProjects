```python
# apply(): allows us to apply this to any function we create
# lambda




import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Courses" : ['MTH101', 'CHM101', 'PHY101', 'TPD101', 'SEA101', 'SEE101', 'SEO101'],
    "Course Coordinators" : ['Mr Fadodun', 'Prof. Ademola', 'Prof. Adeoye', 'Prof Adekunjo', 'Mr Faluyi', 'Dr Ariyo', 'Dr Illori'],
    "Course units" : [5, 4, 4, 1, 2, 2, 2],
    "Exam Date" : ['12 MAY', '5 APRIL', '2 APRIL', '5 APRIL' , '10 JUNE', '12 JUNE', '12 JUNE'],
    "Grades" : [90, 98, 78, 65, 89, 67, 55],
    "Avg population" : [200, 205, 300, 566, 1006, 3444, 78]})
df
```

```python
# The goal is to create the full grades like 90A
# Create a function


def population(x): 
    return format(x * 1000, ",d")
    

# use the apply function
df['Avg population'].apply(func = population)
```

```python
# convert into a df
pd.DataFrame(df['Avg population'].apply(func = population))
```

```python
# using lambda
pd.DataFrame(df['Avg population'].apply(lambda  x: format(x * 100, ",d")))
```
