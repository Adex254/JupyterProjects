```python
# df[['', '', '']].plot(SHIFT + TAB)
# df[[]].hist(SHIFT + TAB)
# The most standard way is
# df[column or list of cols].plot(kind = 'SHIFT + TAB')


import pandas as pd
import numpy as np
import matplotlib as plt


csv_file = pd.read_csv(r"C:\Users\adead\Downloads\csv dataset for practice\diabetes.csv")
df = pd.DataFrame(csv_file)
df
```

```python
# Create the histogram
df['Outcome'].plot(kind = "hist")
```

```python
# Alternnatively
df['Outcome'].hist()
```

```python
# bins
df['Outcome'].hist(bins=5)
# the smaller the bin, the bigger the bars
```

```python
# grid
df['Glucose'].plot(kind = "hist", grid = True)
```

```python
# legend
df['Outcome'].plot(kind = "hist", legend = True)
```

```python
# title
df['Outcome'].plot(kind="hist", title = "My Histiogram Plot", xlabel = "Outcome")
```
