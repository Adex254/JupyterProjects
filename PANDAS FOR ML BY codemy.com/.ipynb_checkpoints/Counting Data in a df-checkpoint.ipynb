```python
import pandas as pd
import numpy as np

excel_file = pd.read_excel(r"C:\Users\adead\Downloads\excel dataset for practice\Financial Sample.xlsx")

df = pd.DataFrame(excel_file)
df.tail(60)
#df.info()
```

```python
# Counting distinct values in desceding order (default) for a specific column
df['Discount Band'].value_counts()
```

```python
# counting in ascending order (by default, ascending = False)
# SHIFT + TAB

df['Discount Band'].value_counts(ascending=True)
```

```python
# counting NaN too. (by default, dropna=True)

df['Discount Band'].value_counts(dropna=False, ascending=True)
```

```python
# percentage or relative frequency (by default, normalize=False)

df['Discount Band'].value_counts(dropna=False, ascending=True, normalize=True)
```

```python
# Using value_counts() for a specific value (rather than a column)
df['Discount Band'].value_counts(dropna=False, ascending=True)['Medium']
```

```python
# Counting unique values - size
df.groupby('Discount Band').size()  # the same as .value_counts()
```

```python
# print out counts by each column
df.groupby('Discount Band').count()
```

```python
# apply()
df.apply(pd.value_counts)
```
