```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


csv_file = pd.read_csv(r"C:\Users\adead\Downloads\csv dataset for practice\diabetes.csv")
df = pd.DataFrame(csv_file)
df.info()
df.describe()
```

```python
# scatter plot
# scatter requires an x and y column

df.plot(kind = 'scatter', x = 'Pregnancies', y = 'Insulin')
plt.show()
```

```python
# third variable c
df.plot(kind = 'scatter', x = 'Pregnancies', y = 'Insulin', c = 'SkinThickness')
plt.show()
```

```python
# cmap 
df.plot(kind = 'scatter', x = 'Pregnancies', y = 'Insulin', c = 'SkinThickness', cmap = 'YlOrBr_r')
plt.show()
```

```python
df.plot(kind = 'scatter', x = 'Pregnancies', y = 'Insulin', c = 'SkinThickness', colormap = 'YlOrBr_r')
plt.show()
```

```python
df.plot(kind = 'scatter', x = 'Pregnancies', y = 'Insulin', c = 'SkinThickness', colormap = 'gist_rainbow_r')
plt.show()
```

```python
df.plot(kind = 'scatter', x = 'Age', y = 'BMI', c = 'DiabetesPedigreeFunction', colormap = 'summer') 
plt.show()
```

```python
df.plot(kind = 'scatter', x = 'Age', y = 'BMI', c = 'DiabetesPedigreeFunction', colormap = 'summer_r', legend = True) 
plt.show()
```
