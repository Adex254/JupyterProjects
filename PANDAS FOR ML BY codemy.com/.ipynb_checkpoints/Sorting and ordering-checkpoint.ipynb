```python
# sort_values()
#  SHIFT + TAB
# use by = ['col 1', 'col 2'] when sorting more than 1 column

import pandas as pd
import numpy as np


df = pd.DataFrame({
    "Courses" : ['MTH101', 'CHM101', 'PHY101', 'TPD101', 'SEA101', 'SEE101', 'SEO101'],
    "Course Coordinators" : ['Mr Fadodun', 'Prof. Ademola', 'Prof. Adeoye', 'Prof Adekunjo', 'Mr Faluyi', 'Dr Ariyo', 'Dr Illori'],
    "Course units" : [5, 4, 4, 1, 2, 2, 2],
    "Exam Date" : ['12 MAY', '5 APRIL', '2 APRIL', '5 APRIL' , '10 JUNE', '12 JUNE', '12 JUNE'],
    "Grades" : [90, 98, 78, 65, 89, 67, 55],
    "Avg population" : [200, 205, 300, 566, 1006, 3444, 78]})
df
```

```python
#  sort Course units - ascending
df['Course units'].sort_values()
# outputs a series

# Alternatively
df.sort_values(by = 'Course units')
```

```python
#  sort many - descending
df[['Avg population', 'Course units', 'Grades']].sort_values(by = ['Avg population', 'Course units', 'Grades'], ascending=False)
```

```python
#  sort many - descending
df.sort_values(by = ['Avg population', 'Course units', 'Grades'], ascending=False)
```
