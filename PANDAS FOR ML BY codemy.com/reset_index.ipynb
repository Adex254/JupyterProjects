```python
import pandas as pd
import numpy as np

csv_file = pd.read_csv(r"C:\Users\adead\Downloads\csv dataset for practice\diabetes.csv")
df = pd.DataFrame(csv_file)
df1 = df.head()   
df1
```

```python
# Create the list of the new row headings
df1['Patients'] = ['Patient 1', 'Patient 2', 'Patient 3', 'Patient 4', 'Patient 5']
df1.set_index(keys= 'Patients')
```

```python
# To modify the DataFrame rather than creating a new one., use inplace=True
df1['Patients'] = ['Patient 1', 'Patient 2', 'Patient 3', 'Patient 4', 'Patient 5']
df1.set_index(keys = 'Patients', inplace=True)
df1
```

```python
df1
```

```python
# To reset index
df1.reset_index(inplace=True)
df1

# The row index goes back to default but our 'Patients' column is still there
```

```python
df1.drop(labels='Patients', axis=1, inplace=True)
df1
```
