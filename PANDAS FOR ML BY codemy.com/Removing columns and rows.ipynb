```python
# Using the  .drop()
# Always designate the axis: axis=0 means the rows and axis=1 means the columns
# inplace=True makes the change permanent
# use SHIFT + TAB  to see  important info the drop()

```

```python
import pandas as pd
import numpy as np

excel_file = pd.read_excel(r"C:\Users\adead\Downloads\excel dataset for practice\Financial Sample.xlsx")
df = pd.DataFrame(excel_file)
df
```

```python
# removing columns
#df.drop('Discount Band', axis=1)   this is only temporarily
```

```python
# To make the change permanent, use inplace=True.
df
df.drop('Discount Band', axis=1, inplace=True)
```
